```python
import numpy as np
import matplotlib.pyplot as plt


time = np.loadtxt('timeLL.txt')
position = np.loadtxt('positionLL.txt')
positionerror = .001
velocityerror = positionerror*2
accelerationerror = positionerror*4

plt.plot(time, position)
plt.plot(time,position,'bo', markersize=2)
plt.xlabel('Time (s)')
plt.ylabel('Position (m) +/- .001 m')
plt.title('Position vs Time')
plt.show()

velocity = (position[2:]-position[:-2])/(2*(time[2:]-time[1:-1]))
time2 = time[1:-1]
plt.plot(time2, velocity)
plt.plot(time2,velocity,'bo', markersize=2)
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s) +/- .002 m/s')
plt.title('Velocity vs Time')
plt.show()

acceleration = (position[2:]-(2*position[1:-1])+position[:-2])/np.square(time[2:]-time[1:-1])

plt.plot(time2, acceleration)
plt.plot(time2,acceleration,'bo', markersize=2)
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s^2) +/- .004 m/s^2')
plt.title('Acceleration vs Time')
plt.show()

print(acceleration)
```
